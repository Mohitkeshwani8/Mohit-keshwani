```python
import requests
import datetime

def fetch_news(keyword):
    now = datetime.datetime.now()
    yesterday = now - datetime.timedelta(hours=100)
    date = now.strftime("%Y-%m-%d")
    api_key = "69a78900e8814a4b9b842d207e275d8b"
    url = "https://newsapi.org/v2/everything?q={}&apiKey={}&from={}&to={}".format(keyword, api_key, yesterday.strftime("%Y-%m-%d"), date)
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data["articles"]
    else:
        return None

news = fetch_news("Assam")
articles_sorted = sorted(news, key=lambda article: article["publishedAt"], reverse=True)

for article in articles_sorted:
    print(article["title"])
    print(article["url"])
    pub_date = datetime.datetime.strptime(article["publishedAt"], "%Y-%m-%dT%H:%M:%SZ")
    print(pub_date.strftime("%B %d, %Y %H:%M:%S"))
```

```python
import tkinter as tk
import requests
import datetime

api_key = "69a78900e8814a4b9b842d207e275d8b"

def fetch_news(keyword):
    now = datetime.datetime.now()
    yesterday = now - datetime.timedelta(hours=8)
    date = now.strftime("%Y-%m-%d")
    url = "https://newsapi.org/v2/everything?q={}&apiKey={}&from={}&to={}".format(keyword, api_key, yesterday.strftime("%Y-%m-%d"), date)
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data["articles"]
    else:
        return None

root = tk.Tk()
root.title("News App")

keyword_entry = tk.Entry(root)
keyword_entry.pack()

def get_news():
    keyword = keyword_entry.get()
    news = fetch_news(keyword)

    articles_frame = tk.Frame(root)
    articles_frame.pack()

    for article in news:
        title = tk.Label(articles_frame, text=article["title"])
        title.pack()
        url = tk.Label(articles_frame, text=article["url"])
        url.pack()

get_news_button = tk.Button(root, text="Get News", command=get_news)
get_news_button.pack()

root.mainloop()
```
